```python
#extract months from a date column and assign weather seasons 
df["date"]=pd.to_datetime(df["date"])
```

```python
#define a function to map months to seasons
def month_to_seasons(month):
    if month in [3,4,5]:
        return "spring"
    elif month in [6,7,8]:
        return "summer"
    elif month in [9,10,11]:
        return "Autumn"
    else:
            return "winter"
```

```python
df["date_month"]=df["date"].dt.month
```

```python
df['season']=df['date_month'].apply(lambda x: month_to_seasons(x))
```

```python
date_seasons=df["season"]
```

```python
print(date_seasons)
```

```python
#Group by weather seasons and calculate avg price
average_price_by_season=df.groupby('season')['price'].mean().reset_index()
```

```python
#visualize the data 
plt.figure(figsize=(8,6))
plt.bar(average_price_by_season['season'],average_price_by_season['price'],color='purple')
plt.title("average price per season")
plt.xlabel("season")
plt.ylabel("average price($)")
plt.show
```
